```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
start_time = time.time()
driver = webdriver.Chrome(executable_path='C://Users//Rithi//Downloads//chromedriver.exe')
driver.get('https://www.cargurus.com/Cars/spt_family_cars')
#titles = [d.text for d in driver.find_elements(By.CLASS_NAME, "soQyMy")]
titles = []
prices = []
miles = []
locations = []
is_new = []
i = 1
while i <= 50:
    for d in driver.find_elements(By.CLASS_NAME, 'soQyMy'):
        for item in d.find_elements(By.CLASS_NAME, 'biZGS4.LtXpwU'):
            #print(item.text)
            titles.append(item.find_element(By.CLASS_NAME, 'vO42pn').text)
            prices.append(item.find_element(By.CLASS_NAME, 'JzvPHo').text)
            miles.append(item.find_element(By.CLASS_NAME, 'SLB6rU').text)
            locations.append(item.find_element(By.CLASS_NAME, 'A7SYzv').text)
            is_new.append(False)

    driver.find_element(By.CLASS_NAME, 'jX_mq2').click()
    time.sleep(3)
    i += 1
end_time = time.time()
total_minutes = (end_time - start_time)/60
print(f"Total Minutes: {total_minutes}")
```

```python
final_prices = [price.rsplit(' ·')[0] for price in prices]
final_prices
```

```python
# Convert the lists in dictionary format
car_dict = {
    'Car_Name': titles,
    'Price': final_prices,
    'Miles': miles,
    'Location': locations,
    'Is_New': is_new
}
```

```python
import pandas as pd
df = pd.DataFrame(car_dict, columns=car_dict.keys())
df.head(25)
```
